```python
# Function to run a game between two AIs
def two_ai_game():
    my_board = make_board()
    while not game_is_over(my_board):
        # The "X" player finds their best move using our custom evaluation function
        result = minimax(my_board, True, 4, -float("Inf"), float("Inf"), my_evaluate_board)
        print("X Turn\nX selected ", result[1])
        select_space(my_board, result[1], "X")
        print_board(my_board)

        if not game_is_over(my_board):
            # The "O" player finds their best move using Codecademy's evaluation function
            result = minimax(my_board, False, 4, -float("Inf"), float("Inf"), codecademy_evaluate_board)
            print("O Turn\nO selected ", result[1])
            select_space(my_board, result[1], "O")
            print_board(my_board)

    # Check and announce the result of the game
    if has_won(my_board, "X"):
        print("X won!")
    elif has_won(my_board, "O"):
        print("O won!")
    else:
        print("It's a tie!")

# A random evaluation function that returns a random score
def random_eval(board):
    return random.randint(-100, 100)

# Our custom evaluation function
def my_evaluate_board(board):
    # Check if "X" has won
    if has_won(board, 'X'):
        return float('Inf')
    # Check if "O" has won
    elif has_won(board, 'O'):
        return -float('Inf')

    # Initialize counters for two-streaks
    x_two_streak, o_two_streak = 0, 0

    # Loop through each column (except the last one)
    for col in range(len(board) - 1):
        # Loop through each row
        for row in range(len(board[0])):
            # Check for "X" two-streaks to the right
            if board[col][row] == 'X' and board[col + 1][row] == 'X':
                x_two_streak += 1
            # Check for "O" two-streaks to the right
            elif board[col][row] == 'O' and board[col + 1][row] == 'O':
                o_two_streak += 1

    # Return the difference between "X" and "O" two-streaks
    return x_two_streak - o_two_streak

# Run a game between two AIs
two_ai_game()

# Example setup to test evaluation function
new_board = make_board()

# Make some moves on the new board
select_space(new_board, 5, "X")
select_space(new_board, 5, "X")
select_space(new_board, 6, "O")
select_space(new_board, 6, "O")
select_space(new_board, 6, "O")
select_space(new_board, 6, "O")
select_space(new_board, 1, "O")
select_space(new_board, 2, "O")

# Uncomment below lines to see the board and evaluation score
# print_board(new_board)
# print(my_evaluate_board(new_board))
```
